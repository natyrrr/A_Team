```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
import plotly.express as px
import os
%matplotlib inline
```

```python
dateCols = ['GRADE DATE']
csvpath= Path("zipcode_restaurants_income.csv", parse_dates=dateCols)
pre_covid_db= pd.read_csv(csvpath)
pre_covid_db.head()
```

```python
Pre_restaurants= pre_covid_db.drop(columns=["CUISINE DESCRIPTION", "SCORE", "GRADE"])
Pre_restaurants.head()
```

```python
Date_pre_db=Pre_restaurants.set_index(Pre_restaurants['GRADE DATE'])
Date_pre_db.head()
```

```python
sorted_pre_restaurants= Date_pre_db.sort_index(ascending= True)
sorted_pre_restaurants.head()
```

```python
pre_rest_db= sorted_pre_restaurants.loc['3/17/2019':'8/06/2019']
pre_rest_db.head()
```

```python
pre_rest_db.duplicated()
```

```python
pre_rest_db= pre_rest_db.drop_duplicates().copy()
```

```python
pre_rest_db['income level'].value_counts()
```

```python
pre_rest_db['New York'].value_counts()
```

```python
pre_rest_db["Borough"].value_counts()
```

```python
# plotting

pre_rest_db["Borough"].value_counts().hvplot.bar(xlabel= 'Borough', ylabel= 'Number of Restaurants', label='Restaurants per Borough- 3/16/19- 8/6/19')
```

```python
pre_rest_db['income level'].value_counts().hvplot.bar(label= "Number of Restaurants in different Socio-Economic Areas, 3/16/19- 8/6/19", xlabel= "Income Bracket", ylabel= " # of Restaurants")
```
